```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

sns.set_theme()
```

```python
df = pd.read_csv("../data/student_habits_performance.csv")
```

```python
df.head()
```

```python
df.info()
```

```python
df.describe()
```

```python
df.isnull().sum()
```

```python
plt.figure(figsize=(10,6))

sns.regplot(
    data=df,
    x="study_hours_per_day",
    y="exam_score",
    scatter_kws={"alpha":0.6}
)

plt.title(
    "Study Hours vs Exam Score",
    fontsize=16,
    fontweight="bold"
)

plt.xlabel("Study Hours Per Day", fontsize=12)
plt.ylabel("Exam Score", fontsize=12)

plt.tight_layout()

plt.savefig(
    "../image/study_vs_score.png",
    dpi=300,
    bbox_inches="tight"
)

plt.show()
```

```python
plt.figure(figsize=(10, 6))

sns.regplot(
    data=df,
    x="social_media_hours",
    y="exam_score",
    scatter_kws={"alpha": 0.6},
)

plt.title("Social Media Usage vs Exam Score", fontsize=16)
plt.xlabel("Hours on Social Media per Day", fontsize=12)
plt.ylabel("Exam Score", fontsize=12)

plt.tight_layout()

plt.savefig("../image/social_media_vs_score.png", dpi=300)

plt.show()
```

```python
plt.figure(figsize=(12,6))

corr = df.select_dtypes(include='number').corr()

sns.heatmap(
    corr,
    annot=True,
    cmap="coolwarm",
    linewidths=0.5
)

plt.title("Correlation Heatmap")

plt.xticks(rotation=45)
plt.yticks(rotation=0)

plt.savefig("../image/heatmap.png")

plt.show()
```

```python
top_students = df.sort_values(
    by="exam_score",
    ascending=False
)

top_students.head(10)
```
